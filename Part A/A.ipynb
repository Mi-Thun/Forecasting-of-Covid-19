###### 6. Which car models have less ‘mpg’ than the first quarter in terms of less ‘mpg’ . Also, find car models that have more ‘mpg’ than the third quarter

```python
# %% Question 6
df2 = df['mpg']
df3 = df['car']
list = []

q1 = float(df2.quantile(.25))
q3 = float(df2.quantile(.75))

def ft(df2, df3, b):
    list.clear()
    c = 0
    for r in df2:
        c += 1
        if b == 0:
            if r < q1:
                list.append(df3[c])
        if b == 1:
            if r > q3:
                list.append(df3[c])
    return list

def show_(a):
    for i in a:
      print(i)

print("Car model Lower than 1st quantile: \n")  # 0 means we want lower
show_(ft(df2, df3, 0))
print("\nCar model Higher than 3rd quantile: \n")  # 1 means we want lower
show_(ft(df2, df3, 1))
```
